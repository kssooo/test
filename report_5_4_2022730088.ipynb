```python
import seaborn as sns

df = sns.load_dataset('iris') # 'iris' dataset을 불러옴.
grid = sns.FacetGrid(data=df, col = 'species') # FacetGrid는 그래프의 '틀'만 제공함. 
                                               # columns 방향으로 'species'에 대한 그래프 틀을 생성함.
    
grid.map(sns.scatterplot, "sepal_length","sepal_width") # FecetGrid는 말 그대로 틀만 만들어주기 때문에 
                                                        # 따로 데이터를 각 틀에 매핑해줌.
                                                        # scatterplot을 사용했기 때문에 점 형태로 데이터 분포가 표현됨.
                                                        # X축, Y축 순서로 데이터 입력을 해주는 형식이기 때문에
                                                        # X축은 sepal_length, Y축은 sepal_width 를 기준으로 하는 분포가 형성됨.
```
